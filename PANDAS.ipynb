```python
##DATA ANALYSIS
#Data Analysis is a method of collecting and organizing the data and if required manipulating the data so that one can 
#derive some useful information from the data.

#Data Analysis and Pandas
#Pandas is a tool in Python that helps you collect (or read) the data from a file, organize it in a tabular format, 
#manipulate and clean the data if required to derive the insightful information from the data.
```

```python
#PANDAS stands for "Python Data Analysis Library"

#It is an open - source Python library
#It is a tool used by data scientists to:
    # 1. read
    # 2. write
    # 3. manipulate, and
    # 4. analyse the data.
    
#Pandas library has two fundamental data structures/objects:
#1. Series
#2. Data Frames

# SERIES:
# This is a one dimensional labeled array. Can hold data of any type and are like a column in a table.
# A Series can have its elements as both numbers and strings but the data type of the Series is always ‘object’ in this case.


##DATAFRAME
# This is a two dimensional table made up of a collection of Series. Structured with labeled axes (rows and columns)
#A DataFrame is a collection of series.

##INDEXING
#Pandas has two known operators used for indexing, they are:
    # 1. loc - label based selection
    # 2. iloc - index based selection
    
#Iloc- Index based selection
# This is to select data based on its numerical position in dataframe.
#The syntax for using iloc operator is "df.iloc[ ]", where df is a dataframe name. 

# loc - Label Based Selection
#Label based selection selects data based on the column or row names/index. 
#loc is used for selecting data based on the data index value/name, not the numerical positions.
#The syntax for loc is similar to iloc: df.loc[ ], df is a dataframe.

##Summary
#1. Count: counts the number of non - null entries in the particular column. 
#2. Unique - the count of unique values in a column. Only for categorical column. 
#3. top - This tells which category is occurring maximum number of times. 
#4. freq - This tells you the number of occurence of the category that is occurring maximum number of times in the column. 
#5. mean - the mean value of the numerical column. 
#6. std - This is standard deviation of the numerical column. This tells you about the variation of data. 
#7. min - the minimum value in the numerical column. 
#8. 25% - the 25th percentile (or 1st quartile) value in the numerical column.
#9. 50% - the 50th percentile (or 2nd quartile or the median) value in the numerical column.
#10. 75% - the 75th percentile (or 3rd quartile) value in the numerical column. 
#11. max - the maximum value in the numerical column. For example, the maximum math score is 100.0
#12. NaN values - means for a particular column a particular summary value is not available. 
```

```python
exam_scores.iloc[0:5, 3]
# Data Indexing using iloc
exam_scores.loc [4:9, "lunch"]
# Data Indexing/ Slicing using loc
exam_scores.gender
# Another method of Data Indexing and Slicing but it can only be used for names without space
exam_scores.rename(columns = {"math score": "math_score", "reading score" : "reading_score", "writing score": "writing_score"},
                  inplace = True)
# Changes the name of a column(s). Note: the inplace function overwrites the renamed columns in the original data.
exam_scores["math_score"]
# Another method of Data Indexing
exam_scores[["math_score", "reading_score", "writing_score"]]
# Selecting multiple columns
exam_scores["reading_score"] == 68
# Returns true if the given argument exists
exam_scores[exam_scores["reading_score"] == 68]
# Returns the given data where is given argument exists
exam_scores["reading_score"] = 99
# Assigns the given argument (99) to the column (reading score)
exam_scores.describe()
# Displays the summary about the numerical columns only
exam_scores.describe (include = "object")
# Displays the summary about the categorical columns only
exam_scores.describe (include = "all")
# Displays the summary about the all columns
exam_scores["parental level of education"].describe()
# Displays the summary about the selected column
exam_scores["test preparation course"].value_counts()
# Counts the unique elements in the colums
exam_scores.gender.unique()
# Displays the uniques objects in the selected column
exam_scores.sort_values(by = "math_score")
# Arranges the data in ascending order of the math score
exam_scores.sort_values (by = "writing_score", ascending = False)
# Arranges the data in descending order of the writing score
exam_scores.columns
# Displays the list of the columns in the DataFrame
exam_scores.index
# Displays the list of the rows (index)in the DataFrame
exam_scores.rename (index = {0 : "zero", 1 : "one"}, inplace = True)
# Changes the name of a row(s)/ index
exam_scores.isnull()
# Returns True for NaN (Not a Number)i.e  for any cell that has no number or value. ."isna()" can also be used.
exam_scores.describe(include = "all").fillna("Missing")
# Fills the empty cells with "Missing"
```
