```python
##MATPLOTLIB
# Matplotlib is the most popular Python plotting library. 
#It is a comprehensive library for creating static, animated, and interactive visualizations in Python.
#It is useful for those working with Python and NumPy.
#It is a low-level library with a Matlab like interface which offers lots of freedom at the cost of having to write more code.

#The most used module of Matplotib is Pyplot which is matplotlib's plotting framework. 
#Each pyplot function makes some change to a figure: e.g., creates a figure, creates a plotting area in a figure, 
#plots some lines in a plotting area, decorates the plot with labels, etc.

# Histogram? 
#A histogram is a graphical display of data using bars(rectangles) of different heights.

## Parts of a Histogram:
# The title: The title describes the information included in the histogram.
# The bars: The height of the bar shows the number of times that the values occurred within the interval, 
# while the width of the bar shows the interval that is covered.
```

```python
import numpy as np

import matplotlib.pyplot as plt

firms = ["Firms A", "Firms B", "Firms C", "Firms D", "Firms E", "Firms F", "Firms G", "Firms H"]
crime_rate = [3, 5, 7, 10, 12, 10, 7, 5]
cent= [1, 3, 5, 7, 9, 11, 13, 15]


plt.scatter(cent, crime_rate)
#plots a scatter diagram of list. (x, y)
plt.show()
# Used to print the diagram. This is used in place of the "print" function.
plt.plot(cent, crime_rate)
# Prints a line graph
plt.hist(crime_rate)
# plots a histogram. It takes only one argument
plt.bar(firms,cent)
# plots a bar chart (vertically)
plt.barh (firms, crime_rate)
# plots a bar chart horizontally
plt.show()
```

```python
Explode =[0, 0, 0, 0, 0.1, 0, 0, 0]
# Explodes the pie chart to emphasis some parts or a certain part
percent = "%1.1f%%"
# Controls how the percentages are displayed in the wedges. One decimal place(1.1), Two decimal places (1.2)
plt.pie (crime_rate, explode = Explode, labels = firms, autopct = percent)
# Displays a pie chart
plt.show()

plt.plot(firms, cent, color = "r", label = "Cent")
# Color = the color the line graph will carry. Label = It's name in the legend
plt.plot(firms, crime_rate, label = "Crime_rate")
# It takes the automatic colour (blue(b))
plt.title ('CENT VS CRIME_RATE COMPARISON')
# Adds title to the graph
plt.xlabel ("FIRMS")
# Adds Label on the x-axis
plt.ylabel ("CENT / CRIME_RATE")
# Adds Label on the y-axis
plt.legend()
# Prints the legend
plt.show()
# Prints a two line graphs in same box
```

```python
#Different methods of Coloring a line graph
    # 1. color = "r"
    # 2. color = "red"
#NOTE: The colors used are RGB (Primary colors) and CMYK (Cyan/Magenta/Yellow/blacK)

# Different methods of changing the line style
    # 1. linestyle = "dashed"
    # 2. linestyle = "--"
    
#NOTE: The linestyle used are = 1. dashed "--" 2. dotted ":" 3. dashdot "-." 4. solid "-"

#Adding markers
    # marker = "^" , "o" , "x"
    # "^" == triangular markers, "o" == circular markers, "x" == "crossed markers"
    
#Linewidth
#This is the width of the marker and linestyle. If no number its indicated den its automatically 1.
```

```python
plt.plot(firms, cent, linestyle = "--", marker = "x", color = "y", label = "Cent", linewidth = 2.5)
plt.legend()
plt.show()

plt.figure(figsize =(8,5))
#Resize the figure (width, height)
plt.plot(firms, crime_rate, ":^g", label = "Crime_rate")
# Another method of writing this. This is a shorter method


plt.show()
```

```python
plt.subplot (1, 2, 1)
# (row, column, index)
plt.plot(firms, cent, "-.co", label = "Cent")
plt.title("Firms vs Cent")

plt.legend()
plt. subplot (1,2,2)
plt.plot (firms, crime_rate, ":m^", label = "Crime_rate")
plt.title ("Firms vs Crime_rate")
plt.suptitle ("FIRMS AGAINST CRIME RATE AND CENT")
# The Title for both graphs

plt.legend()
plt.show()
```
